```python
def busqueda_heuristica(nodo_inicial, solucion, visitados):
    visitados.append(nodo_inicial.get_estado())
    if nodo_inicial.get_estado() == solucion:
        return nodo_inicial
    else:
        # Expandir nodos sucesores (hijos)
        nodo_estado = nodo_inicial.get_estado()
        hijos = generar_hijos(nodo_estado)

        nodo_inicial.set_hijo(hijos)

        for hijo_node in nodo_inicial.get_hijo():
            if not hijo_node.get_estado() in visitados and heuristica(nodo_inicial, hijo_node):
                # Llamada recursiva
                solu = busqueda_heuristica(hijo_node, solucion, visitados)
                if solu is not None:
                    return solu
        return None

def heuristica(padre_node, hijo_node):
    padre_quality = evaluar_calidad(padre_node.get_estado())
    hijo_quality = evaluar_calidad(hijo_node.get_estado())

    return hijo_quality >= padre_quality

def evaluar_calidad(estado):
    calidad = 0
    for n in range(1, len(estado)):
        if estado[n] > estado[n - 1]:
            calidad += 1
    return calidad

def generar_hijos(estado):
    hijos = []
    n = len(estado)
    for i in range(n):
        for j in range(i + 1, n):
            hijo_estado = estado[:]
            hijo_estado[i], hijo_estado[j] = hijo_estado[j], hijo_estado[i]
            hijo_node = NodoBI(hijo_estado)
            hijos.append(hijo_node)
    return hijos
```

```python
import random
if __name__ == "__main__":
    n = 11 # Número de elementos
    bandera = True;
    # estado_inicial = [4, 3, 2, 5, 1]
    estado_inicial = random.sample(range(1, n + 1), n) #random de inicial de rompecabezas
    # estado_solucion = [1, 2, 3, 4, 5]
    estado_solucion = list(range(1, n + 1)) #solucion del randomico
    visitados_nodes = []
    nodo_inicial = NodoBI(estado_inicial)
    nodo_solucion = busqueda_heuristica(nodo_inicial, estado_solucion, visitados_nodes)
    print("Nodo inicial: ",estado_inicial)
    print("Nodo solucion: ",nodo_solucion)
    if nodo_solucion is not None:
        print("Solución encontrada:")
        resultado = []
        node = nodo_solucion
        while node.get_padre() is not None:
            resultado.append(node.get_estado())
            node = node.get_padre()
        resultado.append(estado_inicial)
        resultado.reverse()
        for paso, estado in enumerate(resultado):
            print(f"Paso {paso}: {estado}")
    else:
        print("No se encontró solución.")
```
